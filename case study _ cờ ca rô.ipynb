```python
# B1: tạo của sổ tkinter
import tkinter as tk

root = tk.Tk()
root.resizable(False, False)      # tạo cửa sổ tkinter cố định, không thể thay đổi kích thước
root.title("Cờ ca rô")

# B2: thêm khu vực chơi và label "cờ ca rô"
tk.Label(root, text="Cờ ca rô", font=('Ariel', 25)).pack()  

status_label = tk.Label(root, text="Lượt X", font=('Ariel', 15), bg='green', fg='snow')
status_label.pack(fill=tk.X)

play_area = tk.Frame(root, width=300, height=300, bg='white')  
X_points = []
O_points = [] 
XO_points = []
current_player = 'X'

def play_again():
    global current_player
    current_player = 'X'
    for point in XO_points:
        point.button.configure(state=tk.NORMAL)
        point.reset()
    status_label.configure(text="Lượt X")
    play_again_button.pack_forget()
play_again_button = tk.Button(root, text='Play again', font=('Ariel', 15), command=play_again)

class XOPoint:  
    def __init__(self, x, y):  
        self.x = x  
        self.y = y  
        self.value = None  
        self.button = tk.Button(play_area, text="", width=10, height=5)  
        self.button.grid(row=x, column=y)  
        
      # B3: thêm sự kiện: click nút đổi ký tự và kiểm tra sau mỗi bước X hay O thắng

    def set(self):
        global current_player
        if not self.value:
            self.button.configure(text=current_player, bg='snow', fg='black')
            self.value = current_player
            if current_player == "X":
                X_points.append(self)
                current_player = "O"
                status_label.configure(text="Lượt O")
            elif current_player == "O":
                O_points.append(self)
                current_player = "X"
                status_label.configure(text="Lượt X")
        check_win()
    
    def reset(self):  
        self.button.configure(text="", bg='white')  
        self.value = None  
        
for x in range(1, 4):
    for y in range(1, 4):
        XO_points.append(XOPoint(x, y))    
        
# B4: kiểm tra sau mỗi bước click, X hay O thắng
class WinningPossibility:
    def __init__(self, x1, y1, x2, y2, x3, y3):   # x là hàng, y là cột
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2
        self.x3 = x3
        self.y3 = y3
    def check(self, find):
        p1_satisfied = False
        p2_satisfied = False
        p3_satisfied = False
        if find == 'X':
            for point in X_points:
                if point.x == self.x1 and point.y == self.y1:
                    p1_satisfied = True
                elif point.x == self.x2 and point.y == self.y2:
                    p2_satisfied = True
                elif point.x == self.x3 and point.y == self.y3:
                    p3_satisfied = True
        elif find == 'O':
            for point in O_points:
                if point.x == self.x1 and point.y == self.y1:
                    p1_satisfied = True
                elif point.x == self.x2 and point.y == self.y2:
                    p2_satisfied = True
                elif point.x == self.x3 and point.y == self.y3:
                    p3_satisfied = True
        return all([p1_satisfied, p2_satisfied, p3_satisfied])

winning_possibilities = [
    WinningPossibility(1, 1, 1, 2, 1, 3),
    WinningPossibility(2, 1, 2, 2, 2, 3),
    WinningPossibility(3, 1, 3, 2, 3, 3),
    WinningPossibility(1, 1, 2, 1, 3, 1),
    WinningPossibility(1, 2, 2, 2, 3, 2),
    WinningPossibility(1, 3, 2, 3, 3, 3),
    WinningPossibility(1, 1, 2, 2, 3, 3),
    WinningPossibility(3, 1, 2, 2, 1, 3)
]

for x in range(1, 4):
    for y in range(1, 4):
        XO_points.append(XOPoint(x, y))
def check_win():
    for possibility in winning_possibilities:
        if possibility.check('X'):
            status_label.configure(text="X thắng")
            disable_game()
            return
        elif possibility.check('O'):
            status_label.configure(text="O thắng!")
            disable_game()
            return
    # B5: kiểm tra hòa
    if len(X_points) + len(O_points) == 9:
        status_label.configure(text="hòa!")
        disable_game()

# B6: tạo label thông báo trạng thái (chỉnh lại phía trên)
# B7: hiển thị đến lượt ai (chỉnh lại def set và label trạng thái)
# B8: tạo nút chơi lại


def disable_game():     # vô hiệu hóa game 
    for point in XO_points:
        point.button.configure(state=tk.DISABLED)   
    play_again_button.pack()
        
def play_again():
    global current_player
    current_player = 'x'
    for point in XO_points:
        point.button.configure(state=tk.NORMAL)
        point.reset()
    status_label.configure(text="X's turn")
    play_again_button.pack_forget()
play_again_button = tk.Button(root, text='Play again', font=('Ariel', 15), command=play_again)

def reset(self):
        self.button.configure(text="", bg='lightgray')
        if self.value == "X":
            X_points.remove(self)
        elif self.value == "O":
            O_points.remove(self)
        self.value = None

play_area.pack(pady=10, padx=10) 
root.mainloop()
```
