```python
import turtle
import math
t=turtle.Turtle()
goc = 40
dinh = 100
# điều chỉnh màu bút
t.pencolor('teal')
t.pensize(3)
t.speed(100)
t.penup()
t.backward(150)
t.pendown()
# thiết lập hàm vẽ hình tam giác
def ve_tam_giac_can(a,b,c):
    # a: là góc đáy
    # b: là góc đỉnh
    # c: là cạnh đáy
    canh_ben = c*math.sin(math.radians(a))/math.sin(math.radians(b))
    t.forward(c)
    t.left(180-a)
    t.forward(canh_ben)
    t.left(180-b)
    t.forward(canh_ben)
    t.left(180-a)
# thiết lập hàm vẽ hình chữ nhật
def hinh_chu_nhat(d,r):
    # d: chiều dài
    # r: chiều rộng
    for i in range(2):
        t.forward(d)
        t.right(90)
        t.forward(r)
        t.right(90)
# hàm vẽ tán cây
def tan_cay(a,b,c):
    canh_ben = c*math.sin(math.radians(a))/math.sin(math.radians(b))
    t.right(b/2)
    t.forward(canh_ben)
    t.left(180-a)
    t.forward(c)
    t.left(180-a)
    t.forward(canh_ben)
# vẽ mái nhà
t.fillcolor('red')
t.begin_fill()
ve_tam_giac_can(goc,dinh,150)
t.end_fill()
# vẽ thân ngôi nhà
t.fillcolor('cornsilk')
t.begin_fill()
hinh_chu_nhat(200,100)
t.end_fill()
# đi đến vị trí vẽ ống khói
t.penup()
t.left(goc)
t.forward(150*math.sin(math.radians(goc))/math.sin(math.radians(dinh)))
t.right(180-dinh)
t.forward(40)
t.pendown()
# vẽ ống khói
t.fillcolor('brown')
t.begin_fill()
t.left(130) # dùng góc đồng vị để tính
t.forward(40)
t.right(90)
t.forward(20)
t.right(90)
t.forward(60)
t.end_fill()
# đi đến vị trí vẽ cửa chính
t.penup()
t.goto(0,0)
t.forward(30)
t.right(90)
t.forward(80)
t.right(180)
t.pendown()
# vẽ cửa chính
t.fillcolor('brown')
t.begin_fill()
hinh_chu_nhat(50,70)
t.end_fill()
# đi đến vị trí vẽ cây
t.penup()
t.goto(0,0)
t.left(180)
t.forward(250)
t.left(90)
t.pendown()
# vẽ cây
t.fillcolor('lightgreen')
t.begin_fill()
tan_cay(goc,dinh,50)
t.end_fill()

t.penup()
t.left(130)
t.forward(20.977)    # tính bằng web
t.pendown()
t.fillcolor('darkseagreen')
t.begin_fill()
tan_cay(goc,dinh,70)
t.end_fill()

t.penup()
t.left(130)
t.forward(29.368)    # tính bằng web
t.pendown()
t.fillcolor('forestgreen')
t.begin_fill()
tan_cay(goc,dinh,90)
t.end_fill()
# vẽ thân cây
t.penup()
t.left(130)
t.forward(37.759)  # tính bằng web
t.right(90)
t.forward(10)
t.left(180)
t.pendown()
t.fillcolor('saddlebrown')
t.begin_fill()
hinh_chu_nhat(20,40)
t.end_fill()
# cây thứ 2
t.penup()
t.goto(0,0)
t.forward(200)
t.left(90)
t.forward(50)
t.right(180)
t.pendown()
t.fillcolor('lightgreen')
t.begin_fill()
tan_cay(goc,dinh,100)
t.end_fill()
t.penup()
t.left(130)
t.forward(41.954)    # tính  bằng web
t.pendown()
t.fillcolor('darkseagreen')
t.begin_fill()
tan_cay(goc,dinh,120)
t.end_fill()
t.penup()
t.left(130)
t.forward(50.345)    # tính  bằng web
t.pendown()
t.fillcolor('forestgreen')
t.begin_fill()
tan_cay(goc,dinh,150)
t.end_fill()
# vẽ thân cây
t.penup()
t.left(130)
t.forward(62.932)    # tính  bằng web
t.left(90)
t.backward(20)
t.pendown()
t.fillcolor('saddlebrown')
t.begin_fill()
hinh_chu_nhat(40,80)
t.end_fill()
turtle.done()
```
