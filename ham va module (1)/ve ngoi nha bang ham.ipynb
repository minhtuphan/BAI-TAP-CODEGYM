```python
import turtle
import math
t=turtle.Turtle()
x=100
y=40
# điều chỉnh màu bút
t.pencolor('teal')
t.pensize(3)
# vẽ mái nhà
t.fillcolor('red')
t.begin_fill()
t.backward(x+x/2)
t.left(y)
t.forward(97.906) # tính bằng web
t.right(180-x)
t.forward(97.906)
t.end_fill()
# vẽ thân ngôi nhà
t.fillcolor('cornsilk')
t.begin_fill()
t.left(y)
t.forward(x/2)
t.right(90)
t.forward(x)
t.right(90)
t.forward(2*x)
t.right(90)
t.forward(x)
t.end_fill()
# đi đến vị trí vẽ ống khói
t.penup()
t.right(90-y)
t.forward(97.906)
t.right(180-x)
t.forward(y)
t.pendown()
# vẽ ống khói
t.fillcolor('brown')
t.begin_fill()
t.left(130) # dùng góc đồng vị để tính
t.forward(y)
t.right(90)
t.forward(y/2)
t.right(90)
t.forward(y+y/2)
t.end_fill()
# đi đến vị trí vẽ cửa chính
t.penup()
t.goto(0,0)
t.forward(y-10)
t.right(90)
t.forward(y-10)
t.pendown()
# vẽ cửa chính
t.fillcolor('brown')
t.begin_fill()
t.forward(x/2)
t.left(90)
t.forward(2*y-10)
t.left(90)
t.forward(x/2)
t.left(90)
t.forward(2*y-10)
t.end_fill()
# đi đến vị trí vẽ cây
t.penup()
t.goto(0,0)
t.left(90)
t.forward(2.5*x)
t.pendown()
# vẽ cây
t.fillcolor('lightgreen')
t.begin_fill()
t.left(y)
t.forward(32.635) # tính bằng web
t.left(180-y)
t.forward(x/2)
t.left(180-y)
t.forward(32.635) # tính bằng web
t.end_fill()
t.penup()
t.left(x+30)
t.forward(20.977) # tính bằng web
t.pendown()
t.fillcolor('darkseagreen')
t.begin_fill()
t.right(x/2)
t.forward(45.689) # tính bằng web
t.left(180-y)
t.forward(x/2+20)
t.left(180-y)
t.forward(45.689) # tính bằng web
t.end_fill()
t.penup()
t.left(x+30)
t.forward(29.368) # tính bằng web
t.pendown()
t.fillcolor('forestgreen')
t.begin_fill()
t.right(x/2)
t.forward(58.743) # tính bằng web
t.left(180-y)
t.forward(x-10)
t.left(180-y)
t.forward(58.743) # tính bằng web
t.end_fill()
t.penup()
t.left(x+30)
t.forward(37.759) # tính bằng web
t.right(90)
t.forward(10)
t.left(90)
t.pendown()
t.fillcolor('saddlebrown')
t.begin_fill()
t.forward(y)
t.left(90)
t.forward(y/2)
t.left(90)
t.forward(y)
t.end_fill()
# đi đến vị trí vẽ cây thứ 2
t.penup()
t.right(90)
t.forward(4*x)
t.pendown()
# vẽ thân cây thứ 2
t.fillcolor('saddlebrown')
t.begin_fill()
t.forward(y)
t.right(90)
t.forward(2*y)
t.right(90)
t.forward(y)
t.right(90)
t.forward(2*y)
t.end_fill()
t.right(90)
t.forward(y/2)
t.left(90)
# vẽ tán cây
t.penup()
t.forward(62.932) # tính bằng web
t.pendown()
t.fillcolor('forestgreen')
t.begin_fill()
t.left(x+30)
t.forward(97.906) # tính băng web
t.left(x+y)
t.forward(x+50)
t.left(x+y)
t.forward(97.906) # tính băng web
t.end_fill()
t.penup()
t.right(x/2)
t.forward(50.346) # tính bằng web
t.pendown()
t.fillcolor('darkseagreen')
t.begin_fill()
t.left(x+30)
t.forward(78.324) # tính băng web
t.left(x+y)
t.forward(x+20)
t.left(x+y)
t.forward(78.324) # tính băng web
t.end_fill()
t.penup()
t.right(x/2)
t.forward(41.955) # tính bằng web
t.pendown()
t.fillcolor('lightgreen')
t.begin_fill()
t.left(x+30)
t.forward(65.27) # tính băng web
t.left(x+y)
t.forward(x)
t.left(x+y)
t.forward(65.27) # tính băng web
t.end_fill()
turtle.done()
```
