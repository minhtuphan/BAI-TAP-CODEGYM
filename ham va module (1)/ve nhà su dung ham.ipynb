```python
import turtle
t = turtle.Turtle()
# hàm vẽ hình chữ nhật với a: chiều dài; b: chiều rộng
def ve1(a,b):
    t.color('lightcoral')
    t.begin_fill()
    t.pensize(3)
    t.pencolor('teal')
    t.forward(b)
    t.left(90)
    t.forward(a)
    t.left(90)
    t.forward(b)
    t.left(90)
    t.forward(a)
    t.left(90)
    t.end_fill()
    t.penup()
    t.forward(b/4)
    t.left(90)
    t.forward(a/3)
    t.pendown()
    t.right(90)
    t.color('white')
    t.begin_fill()
    t.pensize(3)
    t.pencolor('teal')
    t.forward(b/3)
    t.left(90)
    t.forward(a/3)
    t.left(90)
    t.forward(b/3)
    t.left(90)
    t.forward(a/3)
    t.left(90)
    t.end_fill()
# vẽ hình vuông
def ve2(a):
    t.color('darkseagreen')
    t.begin_fill()
    t.pensize(3)
    t.pencolor('teal')
    for i in range(4):
        t.forward(a)
        t.right(90)
    t.end_fill()    
    t.penup()
    t.forward(a/3)
    t.right(90)
    t.forward(a/2)
    t.left(90)    
    t.pendown()
    t.color('white')
    t.begin_fill()
    t.pencolor('black')
    t.forward(a/6)
    t.right(90)
    t.forward(a/2)
    t.right(90)
    t.forward(a/6)
    t.right(90)
    t.forward(a/2)
    t.right(90)
    t.forward(a/3)
    t.right(90)
    t.forward(a/2)
    t.right(90)
    t.forward(a/6)
    t.right(90)
    t.forward(a/2)
    t.right(90)
    t.end_fill()    
ve1(150,80)
t.penup()
t.goto(0,0)
t.backward(80)
t.pendown()
ve1(150,80)
t.penup()
t.goto(0,0)
t.right(90)
t.forward(150)
t.left(90)
t.pendown()
ve1(150,80)
t.penup()
t.goto(0,0)
t.right(90)
t.forward(150)
t.left(90)
t.backward(80)
t.pendown()
ve1(150,80)
t.penup()
t.goto(0,0)
t.forward(80)
t.pendown()
ve2(150)
turtle.done()
```
