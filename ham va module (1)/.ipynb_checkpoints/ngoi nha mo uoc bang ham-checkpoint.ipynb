```python
import turtle
import math
t=turtle.Turtle()
goc = 40
dinh = 100
# điều chỉnh màu bút
t.pencolor('teal')
t.pensize(3)
t.speed(0)
t.penup()
t.backward(150)
t.pendown()
# thiết lập hàm vẽ hình tam giác
def ve_tam_giac_can(a,b,c,mau):
    # a: là góc đáy
    # b: là góc đỉnh
    # c: là cạnh đáy
    t.fillcolor(mau)
    t.begin_fill()
    canh_ben = c*math.sin(math.radians(a))/math.sin(math.radians(b))
    t.forward(c)
    t.left(180-a)
    t.forward(canh_ben)
    t.left(180-b)
    t.forward(canh_ben)
    t.left(180-a)
    t.end_fill()
# thiết lập hàm vẽ hình chữ nhật
def hinh_chu_nhat(d,r,mau):
    # d: chiều dài
    # r: chiều rộng
    t.fillcolor(mau)
    t.begin_fill()
    for i in range(2):
        t.forward(d)
        t.right(90)
        t.forward(r)
        t.right(90)
    t.end_fill()
# hàm vẽ tán cây
def tan_cay(a,b,c,mau):
    t.fillcolor(mau)
    t.begin_fill()
    canh_ben = c*math.sin(math.radians(a))/math.sin(math.radians(b))
    t.right(b/2)
    t.forward(canh_ben)
    t.left(180-a)
    t.forward(c)
    t.left(180-a)
    t.forward(canh_ben)
    t.end_fill()
# vẽ mái nhà
ve_tam_giac_can(goc,dinh,150,'red')

# vẽ thân ngôi nhà
hinh_chu_nhat(200,100,'cornsilk')

# đi đến vị trí vẽ ống khói
t.penup()
t.left(goc)
t.forward(150*math.sin(math.radians(goc))/math.sin(math.radians(dinh)))
t.right(180-dinh)
t.forward(40)
t.pendown()
# vẽ ống khói
t.fillcolor('brown')
t.begin_fill()
t.left(130) # dùng góc đồng vị để tính
t.forward(40)
t.right(90)
t.forward(20)
t.right(90)
t.forward(60)
t.end_fill()
# đi đến vị trí vẽ cửa chính
t.penup()
t.goto(0,0)
t.forward(30)
t.right(90)
t.forward(80)
t.right(180)
t.pendown()
# vẽ cửa chính
hinh_chu_nhat(50,70,'brown')

# đi đến vị trí vẽ cây
t.penup()
t.goto(0,0)
t.left(180)
t.forward(250)
t.left(90)
t.pendown()
# vẽ cây
tan_cay(goc,dinh,50,'lightgreen')

t.penup()
t.left(130)
t.forward(20.977)    # tính bằng web
t.pendown()

tan_cay(goc,dinh,70,'darkseagreen')

t.penup()
t.left(130)
t.forward(29.368)    # tính bằng web
t.pendown()

tan_cay(goc,dinh,90,'forestgreen')

# vẽ thân cây
t.penup()
t.left(130)
t.forward(37.759)  # tính bằng web
t.right(90)
t.forward(10)
t.left(180)
t.pendown()

hinh_chu_nhat(20,40,'saddlebrown')

# cây thứ 2
t.penup()
t.goto(0,0)
t.forward(200)
t.left(90)
t.forward(50)
t.right(180)
t.pendown()

tan_cay(goc,dinh,100,'lightgreen')

t.penup()
t.left(130)
t.forward(41.954)    # tính  bằng web
t.pendown()

tan_cay(goc,dinh,120,'darkseagreen')

t.penup()
t.left(130)
t.forward(50.345)    # tính  bằng web
t.pendown()

tan_cay(goc,dinh,150,'forestgreen')

# vẽ thân cây
t.penup()
t.left(130)
t.forward(62.932)    # tính  bằng web
t.left(90)
t.backward(20)
t.pendown()

hinh_chu_nhat(40,80,'saddlebrown')

turtle.done()
```
