```python
danh_sach_sp = {
      101 : {'ten':'SMART WATCH'},
      102 : {'ten': 'PHONE'},
      103 : {'ten': 'PLAYSTATION'},
      104 : {'ten':'LAPTOP'},
      105 : {'ten':'MUSIC PLAYER'},
      106 : {'ten':'TABLET'}   
     }

def them_san_pham(danh_sach_sp):
    print("*** Chức năng thêm sản phẩm ***")
    ma_sp = input("nhập mã sản phẩm: ")
    ten_sp = input("nhập tên sản phẩm: ")
    if ma_sp in danh_sach_sp:
        print(f" sản phẩm có mã {ma_sp} đã có trong danh sach")
    else:
        danh_sach_sp[ma_sp] = {
            'ten': ten_sp
        }
        print(f"*** đã thêm sản phẩm có mã {ma_sp} ***")
        
def sua_san_pham(danh_sach_sp): 
    print("*** Chức năng sửa tên sản phẩm ***")
    ma_sp = input("nhập mã sản phẩm muốn sửa: ")
    if ma_sp not in danh_sach_sp:
        print("sản phẩm không tồn tại")
        return
    ten_sp = input("nhập tên mới cho sản phẩm: ")
    danh_sach_sp[ma_sp] = {
        'ten' : ten_sp
    }
    print(f"*** đã sửa tên sản phẩm có mã {ma_sp} ***")
    
def xoa_san_pham(danh_sach_sp):
    print("*** Chức năng xóa sản phẩm ***")
    ma_sp = input("nhập mã sản phẩm muốn xóa: ")
    if ma_sp not in danh_sach_sp:
        print("sản phẩm không tồn tại")
        return
    danh_sach_sp.pop(ma_sp)
    print(f"*** đã xóa thành công sản phẩm có mã {ma_sp} ***")
    
def hien_thi_danh_sach_sp(danh_sach_sp):
    print("*** Chức năng hiển thị danh sách sản phẩm ***")
    for ma_sp, thong_tin in danh_sach_sp.items():
        print(f"Mã: {ma_sp}, Tên: {thong_tin['ten']}")
while True:
    print("=== ứng dụng quản lý sản phẩm ===")
    print("chọn thao tác:")
    print("1. Thêm sản phẩm")
    print("2. Sửa sản phẩm")
    print("3. Xóa sản phẩm")
    print("4. Hiển thị danh sách sản phẩm")
    print("5. Thoát")
    lua_chon = input("nhập số thao tác: ")
    if lua_chon == '1':
        them_san_pham(danh_sach_sp)
    elif lua_chon == '2':
        sua_san_pham(danh_sach_sp)
    elif lua_chon == '3':
        xoa_san_pham(danh_sach_sp)
    elif lua_chon == '4':
        hien_thi_danh_sach_sp(danh_sach_sp)
    else:
        break
```
