```python
import pandas as pd
import numpy as np
data = pd.read_csv('FoodPrice_in_Turkey.csv',
                  header=0, encoding = 'ISO-8859-1')
data
```

```python
# Việt hóa tên các cột
# Đổi tên chỉ 1 cột xác định
# data.rename(columns = {'Place':'Địa điểm'}, inplace=True)

# Đổi tên chỉ 1 vài cột xác định
# data.rename(columns = {'ProductId':'Mã sản phẩm', 'ProductName':'Tên sản phẩm'}, inplace=True)
# data.rename(columns = {'UmId':'Số lượng', 'UmName':'Tên đơn vị'}, inplace=False) #Bảng data không thay đổi

# Đổi tên toàn bộ các cột (Yêu cầu liệt kê đầy đủ tên cột mới và theo đúng thứ tự)
data.columns = ['Địa điểm', 'Mã SP', 'Tên SP', 'Đơn vị', 'Tên đơn vị', 'Tháng', 'Năm', 'Giá'] #Phải đủ số lượng tên các cột

data
```

```python
# thêm cột mới vào sau cuối
data['tên cột mới'] = np.NaN
data
```

```python
# chèn thêm cột "chiết khấu" vào sau cột 'giá'
data.insert(8,'chiết khấu (%)',20)
display(data)
```

```python
# thêm cột 'số lượng' trước cột 'giá' trá tri 50
data.insert(7,'số lượng',50)
display(data)
```

```python
# đổi 'tên cột mới' thành 'tổng', giá trị bằng giá*số lượng*(100-chiết khấu)
data.rename(columns = {'tên cột mới':'tổng'}, inplace=True)
data['tổng'] = data['Giá'] * data['số lượng'] * (100-data['chiết khấu (%)'])/100
data
```

```python
# thêm dòng mới vào cuối bảng
# chuẩn
data.loc[data.shape[0]]=['Sài Gòn',3000,'Dưa hấu',5,'KG',10,6,2022,0,50.000,0,np.NaN]
display(data)
```

```python
# thêm 1 dòng mới vào đầu bảng
data.loc[-1]=['Hà Nội',3001,'Cốm làng Vòng',5,'KG',10,6,2022,0,50.000,30,np.NaN]
# update index của bảng
data.index = data.index + 1
# sắp xếp lại bẳng
data = data.sort_index()
display(data)
```

```python
# chèn hàng giữa bảng
data.loc[-1] = data.loc[0]
data.loc[0] = ['Đà Nẵng',3002,'Mỳ Quảng',5,'Tô',10,2022,0,20, 100.000,10,np.NaN]
data.index = data.index +1
data = data.sort_index()
display(data)
```
