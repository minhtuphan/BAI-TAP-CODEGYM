```python
import pandas as pd
import numpy as np
data = pd.read_csv('FoodPrice_in_Turkey.csv',
                  header=0, encoding = 'ISO-8859-1')
display(data)
```

```python
# thêm 2 dòng có index -1,2.5 vào bảng
data.loc[-1] = ['ThanhHoa', 66, 'Nem chua', 5, 'KG', 6, 2022, 66.1234]
data.loc[2.5] = ['DaNang', 77, 'BTCTH', 5, 'KG', 6, 2022, 77.1234]
data = data.sort_index()
display(data.head(10))
```

```python
# truy xuất 1 dòng
data.loc[2.5]
```

```python
# dùng iloc
data.iloc[0]
```

```python
# truy xuất nhiều dòng rời rạc
data.loc[[2.5,-1]]
```

```python
# dùng iloc
data.iloc[[4,0]]
```
