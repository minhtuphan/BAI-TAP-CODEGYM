## Đổi tên cột

```python
# Việt hóa tên các cột
# Đổi tên chỉ 1 cột xác định
# data.rename(columns = {'Place':'Địa điểm'}, inplace=True)

# Đổi tên chỉ 1 vài cột xác định
# data.rename(columns = {'ProductId':'Mã sản phẩm', 'ProductName':'Tên sản phẩm'}, inplace=True)
# data.rename(columns = {'UmId':'Số lượng', 'UmName':'Tên đơn vị'}, inplace=False) #Bảng data không thay đổi

# Đổi tên toàn bộ các cột (Yêu cầu liệt kê đầy đủ tên cột mới và theo đúng thứ tự)
data.columns = ['Địa điểm', 'Mã SP', 'Tên SP', 'Đơn vị', 'Tên đơn vị', 'Tháng', 'Năm', 'Giá'] #Phải đủ số lượng tên các cột

data
```

## Thêm cột vào bảng đã có

```python
# Thêm cột mới vào sau cuối
data['tên cột mới'] = np.NaN
data
```

```python
# Chèn thêm cột 'Chiết khấu' vào sau cột 'Giá'
data.insert(8,'Chiết khấu (%)', 20)
display(data)
# data.info()
```

```python
# Thêm 1 cột tên là "số lượng" nằm trước cột Giá, giá trị toàn bộ là 50
data.insert(7,'Số lượng',50)
data
```

```python
# Đổi tên "Tên cột mới" thành "Tổng" và có giá trị là giá*số lượng*(100%-chiết khấu)
# gợi ý: data['A'] = data[B]*C*(100-D)/100
data.rename(columns = {'tên cột mới':'tổng'}, inplace=True)
data['tổng'] = data['Giá'] * data['Số lượng'] * (100-data['Chiết khấu (%)'])/100
data
```

### Thêm 1 dòng mới vào đầu bảng

```python
# Thêm 1 dòng mới vào đầu bảng
data.loc[-1]= ['Hà Nội',3001,'Cốm làng Vòng',5,'KG',10,2022,20, 50.000,30,np.NaN]
# Update index của bảng
data.index = data.index + 1
# Sắp xếp lại bảng
data = data.sort_index()
display(data)
# data.info()
```

### Thêm 1 dòng mới vào vị trí bất kỳ

```python
#Cách 1: Swap giá trị các dòng
data.loc[-1] = data.loc[0]
data.loc[0] = ['Đà Nẵng',3002,'Mỳ Quảng',5,'Tô',10,2022,20, 100.000,10,np.NaN]
data.index = data.index +1
data = data.sort_index()
```

```python
# Cách 2: Chèn thêm ở vị trí trung gian
# Chèn thêm dòng ở vị trí bất kỳ, ví dụ 0.5
data.loc[0.5]= ['Đà Nẫng',3002,'Mì quảng',5,'tô',10,9,2022,100.000,20,np.NaN]

#Sắp xếp lại các hàng theo đúng thứ tự từ nhỏ tới lớn và theo số tự nhiên
data = data.sort_index().reset_index(drop=True)

display(data)
```

## Xóa cột

## Xóa cột sử dụng lệnh del
### del df['Tên cột cần xóa']

```python
# Xóa cột sử dụng del
data2 = data
print('Data2 sau khi tạo')
display(data2)
# Xóa cột 'cột mới' (xóa ngay trên bảng gốc) -> Nguy hiểm, hạn chế dùng
del data2['tổng']
print('Data2 sau khi xóa')
display(data2)
print('Data sau khi xóa')
display(data)
```

## Xóa cột sử dụng hàm pop()
### df.pop('Tên cột cần xóa')

```python
# Xóa cột sử dụng pop()
data3 = data
print('Data3 sau khi tạo')
display(data3)
# Xóa cột 'cột mới' (xóa ngay trên bảng gốc) -> hạn chế dùng
data4 = data3.pop('Chiết khấu (%)') #-> kết quả là cột cần xóa
print('Data3 sau khi xóa')
display(data3)
print('Data sau khi xóa')
display(data)
print('Data4')
display(data4)
```

## Xóa cột sử dụng hàm drop()
### df.drop('Tên cột cần xóa’, axis=1, inplace=True/False)
axis = 1 nghĩa là xóa cột
inplace = True -> xóa cả dữ liệu gốc = del, pop()
inplace = False -> dữ lại nguyên vẹn dữ liệu gốc (mặc định)

```python
# Xóa cột sử dụng drop()
data4 = data
print('Data4')
display(data4)
# Xóa cột 'cột mới' (không xóa trực tiếp trên bảng gốc) -> nên dùng
data5 = data4.drop('Số lượng', axis=1, inplace=False)  #-> lấy kết quả trả về
# data5 = data4.drop('Số lượng', axis=1, inplace=True)  #-> không trả về kết quả mà cột bị xóa trực tiếp trên bảng gốc
print('Data4 sau khi xóa')
display(data4)
print('Data5')
display(data5)
print('Data gốc')
display(data)
```

## Xóa nhiều cột sử dụng hàm drop()
### df.drop(['Tên cột cần xóa1',…, 'Tên cột cần xóa n'], axis=1, inplace=True/False)

```python
# Xóa cột 'cột mới' và cột 'số lượng'
data6 = data.drop(['Chiết khấu (%)', 'Số lượng'], axis=1)
print('Data6')
display(data6)
print('Data gốc')
display(data)
```

## Xóa dòng sử dụng hàm drop()
### df.drop(cs, axis=0, inplace=True/False)
axis = 0 -> xóa dòng
cs là index của dòng đấy

```python
#Xóa dòng cuối cùng
data6 = data.drop(data.shape[0]-1, axis=0)
print("Data6")
display(data6)
```

```python
# Xóa dòng đầu tiên của bảng
data7 = data.drop(0, axis=0) #-> xóa đi và không reset lại index
display(data)
display(data7)
# data7.info()
```
