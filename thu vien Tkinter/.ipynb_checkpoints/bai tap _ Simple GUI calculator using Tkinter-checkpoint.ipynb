```python
import tkinter as tk
from tkinter import ttk

main = tk.Tk()
main.title('simple GUI calculator')
expression = ""
def update_expression_input(exp):
    expression_input.delete(0, tk.END)
    expression_input.insert(0, exp)
    
def handle_click(value):
    global expression
    expression += value
    update_expression_input(expression)
    
def handle_click_equal():
    global expression
    try:
        update_expression_input(eval(expression))
    except:
        update_expression_input('error')
    expression = ''
    
def handle_click_clear():
    global expression
    update_expression_input('')
    expression = ''
    
expression_input = ttk.Entry(main)
expression_input.grid(row=0, column=0, columnspan=4, sticky='we')

button_1 = ttk.Button(main, text='1', command=lambda: handle_click('1'))
button_1.grid(row=1, column=0)

button_2 = ttk.Button(main, text='2',command=lambda: handle_click('2'))
button_2.grid(row=1, column=1)

button_3 = ttk.Button(main, text='3',command=lambda: handle_click('3'))
button_3.grid(row=1, column=2)

button_plus = ttk.Button(main, text='+',command=lambda: handle_click('+'))
button_plus.grid(row=1, column=3)

button_4 = ttk.Button(main, text='4',command=lambda: handle_click('4'))
button_4.grid(row=2, column=0)

button_5 = ttk.Button(main, text='5',command=lambda: handle_click('5'))
button_5.grid(row=2, column=1)

button_6 = ttk.Button(main, text='6',command=lambda: handle_click('6'))
button_6.grid(row=2, column=2)

button_minus = ttk.Button(main, text='-',command=lambda: handle_click('-'))
button_minus.grid(row=2, column=3)

button_7 = ttk.Button(main, text='7',command=lambda: handle_click('7'))
button_7.grid(row=3, column=0)

button_8 = ttk.Button(main, text='8',command=lambda: handle_click('8'))
button_8.grid(row=3, column=1)

button_9 = ttk.Button(main, text='9',command=lambda: handle_click('9'))
button_9.grid(row=3, column=2)

button_multiply = ttk.Button(main, text='*',command=lambda: handle_click('*'))
button_multiply.grid(row=3, column=3)

button_0 = ttk.Button(main, text='0',command=lambda: handle_click('0'))
button_0.grid(row=4, column=0)

button_clear = ttk.Button(main, text='clear',command=handle_click_clear)
button_clear.grid(row=4, column=1)

button_equal = ttk.Button(main, text='=',command=handle_click_equal)
button_equal.grid(row=4, column=2)

button_devide = ttk.Button(main, text='/',command=lambda: handle_click('/'))
button_devide.grid(row=4, column=3)

button_dot = ttk.Button(main, text='.',command=lambda: handle_click('.'))
button_dot.grid(row=5, column=0)

main.mainloop()
```
